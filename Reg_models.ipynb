```python
import altair as alt


plt1 = alt.Chart(dataFr).mark_circle(size=60, color='#ca5492').encode(x =alt.X('bug', axis=alt.Axis(title='wheat')), y = 'Y', tooltip = 'Name:N').interactive()
plt2 = alt.Chart(dataFr).mark_circle(size=60, color='purple').encode(x =alt.X('ar', axis=alt.Axis(title='don')), y = 'Y',).interactive()
plt3 = alt.Chart(dataFr).mark_circle(size=60, color='orange').encode(x ='mak', y = 'Y',).interactive()
plt4 = alt.Chart(dataFr).mark_circle(size=60, color='green').encode(x ='sholi', y = 'Y',).interactive()
plt5 = alt.Chart(dataFr).mark_circle(size=60, color='red').encode(x ='duk', y = 'Y',).interactive()

plt1 + plt2 + plt3 + plt4 + plt5
```
